```
img = cv2.imread('sideview.png')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)

lChannel = gray[:,:,1]
mask = cv2.inRange(lChannel,170,255)
res = cv2.bitwise_and(img,img,mask=mask)
cv2_imshow(res)
```

```
blur = cv2.GaussianBlur(res,(5,5),0)
cv2_imshow(blur)
```

```
edge = cv2.Canny(blur,100,200)
cv2_imshow(edge)
```
